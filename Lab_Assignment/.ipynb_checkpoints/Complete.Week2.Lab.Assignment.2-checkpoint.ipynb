```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("data2.sqlite")  # Ensure the correct path to your database file

# Define the SQL query using a subquery instead of a JOIN
query = """
SELECT customerNumber, 
       contactLastName, 
       contactFirstName
FROM customers
WHERE customerNumber IN (
    SELECT customerNumber 
    FROM orders 
    WHERE orderDate = '2003-01-31'
);
"""

# Execute the query and store the results in a DataFrame
df = pd.read_sql_query(query, conn)

# Display the results
df
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect("data2.sqlite")  # Ensure this path is correct

# SQL Query to count total orders per product name
query_q2 = """
SELECT productName, COUNT(*) AS total_orders
FROM products
JOIN orderdetails USING(productCode)
GROUP BY productName
ORDER BY total_orders DESC;
"""

# Execute query and store results in DataFrame
df_q2 = pd.read_sql_query(query_q2, conn)

# Display the results
df_q2
```

```python
# SQL Query to count unique customers per product
query_q3 = """
SELECT productName, COUNT(DISTINCT customerNumber) AS total_customers
FROM products
JOIN orderdetails USING(productCode)
JOIN orders USING(orderNumber)
GROUP BY productName
ORDER BY total_customers DESC;
"""

# Execute query and store results in DataFrame
df_q3 = pd.read_sql_query(query_q3, conn)

# Display the results
df_q3
```

```python
query_q4 = """
SELECT DISTINCT e.employeeNumber, e.firstName, e.lastName, o.city, e.officeCode
FROM employees e
JOIN customers c ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders o ON c.customerNumber = o.customerNumber
JOIN orderdetails od ON o.orderNumber = od.orderNumber
JOIN products p ON od.productCode = p.productCode
JOIN offices o ON e.officeCode = o.officeCode  -- Explicitly joining offices
WHERE p.productCode IN (
    SELECT productCode
    FROM orderdetails
    JOIN orders USING(orderNumber)
    GROUP BY productCode
    HAVING COUNT(DISTINCT customerNumber) < 20
);
"""

df_q4 = pd.read_sql_query(query_q4, conn)
df_q4

```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("data2.sqlite")  # Ensure this path is correct

# SQL Query to select employees whose customers have an avg. credit limit > 15K
query_q5 = """
SELECT e.employeeNumber, e.firstName, e.lastName, COUNT(c.customerNumber) AS total_customers
FROM employees e
JOIN customers c ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY e.employeeNumber
HAVING AVG(c.creditLimit) > 15000
ORDER BY total_customers DESC;
"""

# Execute query and store results in DataFrame
df_q5 = pd.read_sql_query(query_q5, conn)

# Display the results
df_q5
```
